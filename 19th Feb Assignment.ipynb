### Q1. What is Flask Framework? What are the advantages of Flask Framework?

#### Ans:
Flask is a lightweight web framework written in Python that is designed to make it easy to develop web applications quickly and with minimal setup. It is known for its simplicity and flexibility, which make it a popular choice for building small to medium-sized web applications.

Advantages of Flask Framework:
    
1. Lightweight: Flask is a lightweight framework that doesn't come with any pre-built components, which makes it easy to install and use.

2. Easy to learn: Flask has a simple and intuitive syntax that makes it easy to learn for developers of all skill levels.

3. Highly customizable: Flask provides a lot of flexibility in terms of customization, allowing developers to build web applications that meet their specific needs.

4. Extensible: Flask is extensible, which means that developers can add new functionality to the framework by installing and using third-party extensions.

5. Large community: Flask has a large and active community of developers who provide support and contribute to the development of the framework.

6. Built-in development server: Flask comes with a built-in development server, which makes it easy to test and debug applications during development.

7. Compatibility: Flask is compatible with a wide range of databases, including MySQL, PostgreSQL, and SQLite.

### Q3. What is App routing in Flask? Why do we use app routes?

#### Ans:
In Flask, app routing refers to the process of defining URL routes for our web application. When a client sends a request to a Flask application, the request is matched against the defined URL routes to determine which view function should handle the request.

We use app routes in Flask to define the URLs that our application will respond to, and to map those URLs to specific view functions. This makes it easy to create a well-organized and maintainable web application, since each view function can be responsible for handling a specific URL route and related functionality.

### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the url_for() function is used for URL building. This function generates a URL for the given endpoint and any arguments passed to it, based on the defined routes in the Flask application.

The url_for() function takes the endpoint name as its first argument, followed by any keyword arguments that correspond to the variable parts of the route pattern. When called with just the endpoint name as the argument, url_for() will return a URL that corresponds to the route pattern for that endpoint without any arguments.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/hello')
def hello():
    return 'Hello, World!'

@app.route('/user/<username>')
def show_user_profile(username):
    return 'User %s' % username

@app.route('/post/<int:post_id>')
def show_post(post_id):
    return 'Post {}'.format(post_id)

if __name__ == '__main__':
    with app.test_request_context():

        print(url_for('hello'))

        print(url_for('show_user_profile', username='XYZ'))

        print(url_for('show_post', post_id=1))
```
